```python
import pandas as pd

parquet_path = r"test.parquet"
df = pd.read_parquet(parquet_path, engine="pyarrow")

df.head()
```

```python
df.columns
```

```python
# Chr
if "CHROM_norm" in df.columns:
    df["Chr"] = (
        df["CHROM_norm"]
        .astype(str)
        .str.replace("chr", "", regex=False)
    )
elif "CHROM" in df.columns:
    df["Chr"] = (
        df["CHROM"]
        .astype(str)
        .str.replace("chr", "", regex=False)
    )
else:
    raise ValueError("❌ Không tìm thấy CHROM hoặc CHROM_norm")

# Position
if "POS" not in df.columns:
    raise ValueError("❌ Thiếu cột POS")

df["Start"] = df["POS"]
df["End"] = df["POS"]  # SNV → Start = End

# REF - ALT
if "REF" not in df.columns or "ALT" not in df.columns:
    raise ValueError("❌ Thiếu REF hoặc ALT")

df["Ref"] = df["REF"]
df["Alt"] = df["ALT"]


df_magpie = df[["Chr", "Start", "End", "Ref", "Alt"]].copy()

# Save
output_path = "magpie_input.csv"
df_magpie.to_csv(output_path, index=False)

print("MAGPIE minimal input saved to:", output_path)
print(df_magpie.head())
print("Shape:", df_magpie.shape)
```
