```python
# Danh sách giá trị cần kiểm tra
invalid_values = ["-", "na", "n/a", "NA", "N/A", "NaN", "nan"]

# Thay thế toàn bộ giá trị lỗi trong toàn DataFrame bằng np.nan
dff = dff.replace(invalid_values, np.nan)
```

```python
dff.isnull().sum()
```

```python
dff = dff.rename(columns={"Chromosome": "CHROM"})
dff = dff.rename(columns={"PositionVCF": "POS"})
dff = dff.rename(columns={"ReferenceAlleleVCF": "REF"})
dff = dff.rename(columns={"AlternateAlleleVCF": "ALT"})
def normalize_chromosome(chrom_name):
    chrom_str = str(chrom_name).strip() # Chuyển sang string và xóa khoảng trắng

# Lọc bỏ các contig không xác định (Unplaced/random/alt)
    if 'Un' in chrom_str or 'random' in chrom_str or 'alt' in chrom_str:
        return None # Sẽ bị xóa ở bước tiếp theo

    # Xử lý các trường hợp đặc biệt (MT -> chrM)
    if chrom_str.upper() == 'MT':
        return 'chrM'

    # Nếu nó đã có 'chr' rồi thì thôi (như chr1, chrX, chrM của bạn)
    if chrom_str.startswith('chr'):
        return chrom_str

    # Nếu nó là số (1, 2, X, Y) thì thêm 'chr'
    return 'chr' + chrom_str

print("Đang chuẩn hóa tên chromosome ...")
dff['CHROM'] = dff['CHROM'].apply(normalize_chromosome)

original_count = len(dff)
dff = dff.dropna(subset=['CHROM']) # Xóa các hàng (như chrUn) đã bị gán None
dff = dff.dropna(subset=['POS'])
dff = dff.dropna(subset=['REF'])
dff = dff.dropna(subset=['ALT'])
removed_count = original_count - len(dff)
print(f"Đã lọc bỏ {removed_count} hàng ('chrUn' hoặc contig rác, thiếu POS, ALT, REF).")

print("Đang sắp xếp dff (theo chrom và pos)...")
# 1. Đảm bảo 'pos' là kiểu số (integer) để sắp xếp đúng
dff['POS'] = pd.to_numeric(dff['POS'])
# 2. Sắp xếp theo 'chrom' trước, sau đó theo 'pos'
dff = dff.sort_values(by=["CHROM", "POS"])
dff
```
